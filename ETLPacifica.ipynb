```python
#Data transformation 

# Datetime conversion

df['Job In Date'] = pd.to_datetime(df['Job In Date'], dayfirst=True)
df['Placement Date'] = pd.to_datetime(df['Placement Date'], dayfirst=True)
df['Job Out Date'] = pd.to_datetime(df['Job Out Date'], dayfirst=True)
df['Forecast']=pd.to_datetime(df['Forecast'], dayfirst=True)

# Rename column 'Monhly Salary' to 'Monthly Salary'
df = df.rename(columns=lambda x: x.strip())
df = df.rename(columns={'Monhly Salary': 'Monthly Salary'})

# Verificar que la columna 'Monthly Salary' exista
if 'Monthly Salary' in df.columns:
    # Convertir los valores a cadenas para usar str.replace()
    df['Monthly Salary'] = df['Monthly Salary'].astype(str)
    
    # Reemplazar puntos y comas
    df['Monthly Salary'] = df['Monthly Salary'].str.replace('.', '', regex=False)
    df['Monthly Salary'] = df['Monthly Salary'].str.replace(',', '.', regex=False)
    
    # Convertir a float
    df['Monthly Salary'] = df['Monthly Salary'].astype(float)
    
    # Redondear y convertir a int
    df['Monthly Salary'] = df['Monthly Salary'].round(0).astype(int)


# Transform 'Anual Salary' to float and then to integer
if 'Anual Salary' in df.columns:
    # Convertir los valores a cadenas para usar str.replace()
    df['Anual Salary'] = df['Anual Salary'].astype(str)
    
    # Reemplazar puntos y comas
    df['Anual Salary'] = df['Anual Salary'].str.replace('.', '', regex=False)
    df['Anual Salary'] = df['Anual Salary'].str.replace(',', '.', regex=False)
    
    # Convertir a float
    df['Anual Salary'] = df['Anual Salary'].astype(float)
    
    # Redondear y convertir a int
    df['Anual Salary'] = df['Anual Salary'].round(0).astype(int)

df['Fee%'] = df['Fee%'].str.rstrip('%').astype('float')


if 'Local Fee' in df.columns:
    # Eliminar espacios en blanco alrededor de los valores
    df['Local Fee'] = df['Local Fee'].str.strip()

    # Reemplazar puntos con nada y comas con puntos
    df['Local Fee'] = df['Local Fee'].str.replace('.', '', regex=False)
    df['Local Fee'] = df['Local Fee'].str.replace(',', '.', regex=False)
    
    # Reemplazar valores no válidos (por ejemplo, ' - ') con NaN
    df['Local Fee'] = df['Local Fee'].replace(['-', ''], np.nan)
    
    # Convertir la columna a float
    df['Local Fee'] = df['Local Fee'].astype(float)
```
